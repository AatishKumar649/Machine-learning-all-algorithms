# **Linear Regression**

```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from pandas import read_csv
from google.colab import files

# Upload housing.csv from your local machine
data=read_csv("/content/housing.csv")

# Column names
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM',
                'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B',
                'LSTAT', 'MEDV']

# Load dataset directly after upload
data = read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
print(data.head(5))
```

```python
# Independent features (X) and dependent feature (y)
X = data.drop('MEDV', axis=1)   # all columns except MEDV
y = data['MEDV']                # only MEDV column

# Print target variable
print("Dependent Feature (y):")
print(y.head(10))
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into train (80%) and test (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
X_train
```

```python
## standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
##cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
##prediction
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
score=r2_score(reg_pred,y_test)
score
```
